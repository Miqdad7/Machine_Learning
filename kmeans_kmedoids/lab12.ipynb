```python
import pandas as pd
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import rand_score,adjusted_rand_score,silhouette_score
```

```python
iris_df=pd.read_csv('Iris.csv')
feature,target=make_blobs(n_samples=500,centers=5,random_state=42,shuffle=False)
plt.scatter(feature[:,0],feature[:,1])
```

```python
model=KMeans(n_clusters=4)
model.fit(feature)
plt.scatter(feature[:,0],feature[:,1],color="r",alpha=0.5,label="Data Points")

plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],marker="*",s=200,color="k",label="Cluster centers")

plt.title("KMeans Clustering")
plt.xlabel("Featurre 1")
plt.ylabel("feature 2")
plt.legend()
plt.show()
```

```python
rs=rand_score(target,model.labels_)
ars=adjusted_rand_score(target,model.labels_)
sil=silhouette_score(feature,model.labels_)

print(rs)
print(ars)
print(sil)
```
