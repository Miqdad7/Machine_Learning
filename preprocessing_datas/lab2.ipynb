```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
fifa_df_complete=pd.read_csv('fifa23.csv')
fifa_df=fifa_df_complete.iloc[:51].copy()
countries=np.unique(fifa_df['Nationality'])

cle=LabelEncoder()
country_labels=cle.fit_transform(fifa_df['Nationality'])
fifa_df['Country_Labels']=country_labels

print(fifa_df[['Known As','Age','Overall','Nationality','Country_Labels']].head(10))

country_mappings={}
for index,label in enumerate(cle.classes_):
    country_mappings[index]=label
print(country_mappings)
```

```python
ale=LabelEncoder()
age_labels=ale.fit_transform(fifa_df['Age'])
fifa_df['Age Label']=age_labels
print(fifa_df[['Known As','Age','Age Label','Overall','Nationality','Country_Labels']].head(10))

age_mappings={}
for index,label in enumerate(ale.classes_):
    age_mappings[index]=label
print(age_mappings)
```

```python
pos_onehot_features=pd.get_dummies(fifa_df['Best Position'])
pos_onehot_features=pos_onehot_features.astype(int)
pd.concat([fifa_df[['Known As','Age','Overall','Nationality','Country_Labels']],pos_onehot_features],axis=1).head(10)
```
