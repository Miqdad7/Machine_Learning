```python
import random
import datetime
import pandas as pd

no_of_records=30
dates=pd.date_range(start='2023-01-01',end='2023-12-31',periods=no_of_records).to_pydatetime()
dates=[d.date() for d in dates]
prices=[round(random.uniform(10,100),2) for _ in range(no_of_records)]
product_ids=[random.randint(1,100) for _ in range(no_of_records)]
quantities=[random.randint(1,10) for _ in range(no_of_records)]
serial_nos=[f'SN-{random.randint(1000,9999)}' for _ in range(no_of_records)]
user_ids=[f'U-{random.randint(10,20)}' for _ in range(no_of_records)]
user_types=['Retail','Wholesale']
user_classes=['Class A','Class B','Class C']
purchase_weeks=[d.isocalendar()[1] for d in dates]
data={
    'Date':dates,
    'Price':prices,
    'Product_ID':product_ids,
    'Quantity_Purchased':quantities,
    'Serial_No':serial_nos,
    'User_ID':user_ids,
    'User_Type':random.choices(user_types,k=no_of_records),
    'User_Class':random.choices(user_classes,k=no_of_records),
    'Purchase_Week':purchase_weeks
}
df=pd.DataFrame(data)
df.to_csv('customer_dataset.csv',index=False)
```

```python
import matplotlib.pyplot as plt
df=pd.read_csv('customer_dataset.csv')
user_id='U-20'
user_data=df[df['User_ID']==user_id]

plt.figure(figsize=(10,6))
plt.plot(user_data['Date'],user_data['Price'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'Price trends for User ID: {user_id}')
plt.xticks(rotation=90)
```

```python
df=pd.read_csv('customer_dataset.csv')
user_id='U-20'
user_data=df[df['User_ID']==user_id]
plt.figure(figsize=(10,6))
plt.plot(df['Date'],df['Price'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'Price trends for User ID: {user_id}')
plt.xticks(rotation=90)
plt.grid()
plt.scatter(user_data['Date'],user_data['Price'],color='red',label=f'User ID: {user_id}')
plt.legend(loc='upper left')

plt.show()
```
